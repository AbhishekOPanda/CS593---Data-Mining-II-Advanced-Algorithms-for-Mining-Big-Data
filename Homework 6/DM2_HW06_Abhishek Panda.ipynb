```python
# Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
# Load dataset
data = pd.read_csv('depression.csv')
data
```

```python
# Input
X = data[["SEX", "AGE", "INCOME","BEDDAYS","HEALTH"]]

# Output
y = data["Cat_total"]
```

```python
# Split the data 
xtrn, xtst, y_trn, y_tst = train_test_split(X, y, test_size = 0.25, random_state = 0)
xtrn
```

```python
sc_x = StandardScaler()
xtrn = sc_x.fit_transform(xtrn)
xtst = sc_x.transform(xtst)
```

```python
# Define Logistic Rgression Model
classifier = LogisticRegression(random_state = 123)
classifier.fit(xtrn, y_trn)
```

```python
y_pred = classifier.predict(xtst)
y_pred
```

```python
y_pred = ["Yes" if year > 15 else "No" for year in y_pred]
y_tst = ["Yes" if year > 15 else "No" for year in y_tst]
```

```python
# Accuracy
print ("Accuracy : ", accuracy_score(y_tst, y_pred))
```

```python
# Confusion Matrix
cm = confusion_matrix(y_tst, y_pred)
print ("Confusion Matrix : \n", cm)
```
